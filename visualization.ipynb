```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.style.use('seaborn-pastel')
```

```python
np.random.seed(123)
df=pd.DataFrame({'A':np.random.randn(365).cumsum(0),'B':np.random.randn(365).cumsum(0)+20,'C':np.random.randn(365).cumsum(0)-20})
```

```python
df
```

```python
df.plot()
```

```python
plt.style.use('seaborn-colorblind')
```

```python
df.plot()
```

```python
plt.figure(figsize=(10,10))
plt.plot(df,label='x')
plt.xlabel('Random')
plt.ylabel('Random')
plt.title('Random')
plt.legend(loc='best',bbox_to_anchor=(1,1))
```

```python
#SPLOMS
```

```python
cols=['A','B','C']
```

```python
fig,ax=plt.subplots(len(cols),len(cols))
for i in range(len(cols)):
    for j in range(len(cols)):
        ax[i,j].scatter(df[cols[i]],df[cols[j]],c='r',s=5)
        ax[i,j].get_xaxis().set_visible(False)
        ax[i,j].get_yaxis().set_visible(False)
        if i==len(cols)-1:
            ax[i,j].get_xaxis().set_visible(True)
        if j==0:
            ax[i,j].get_yaxis().set_visible(True)
```

```python
v1=pd.Series(np.random.normal(0,10,1000),name='v1')
v2=pd.Series(np.random.normal(60,15,1000)+2*v1,name='v2')
```

```python
v1
```

```python
v1.hist(bins=np.arange(-50,50,5))
```

```python
plt.hist(v1,bins=np.arange(-50,50,5))
plt.grid(which='major')
```

```python
v2.hist(bins=np.arange(-50,150,5),width=0.9)
sns.kdeplot(v2)
```

```python
v3=np.concatenate((v1,v2))
```

```python
plt.hist([v1,v2],density=True)
sns.kdeplot(v3)
```

```python
sns.histplot(v3,color='Teal',kde=True)
```

```python
sns.pairplot(df,diag_kind='kde')
```

```python
sns.jointplot(x=v1,y=v2,space=0,kind='kde')
```
